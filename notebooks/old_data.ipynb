```python
import sqlite3
from pathlib import Path

import pandas as pd
```

```python
DATA_DIR = Path('..') / 'data'

DB = DATA_DIR / 'angiosperms.sqlite'
```

```python
sql = """
    select *
    from angiosperms
    join images using (coreid)
"""
with sqlite3.connect(DB) as cxn:
    df = pd.read_sql(sql, cxn)
df.shape
```

```python
df["name"] = df["path"].str.split("/").str[-1]
df.head()
```

```python
ant0 = pd.read_csv(DATA_DIR / "ant" / "ant_0.csv")
ant1 = pd.read_csv(DATA_DIR / "ant" / "ant_1.csv")
ant = pd.concat([ant0, ant1])
ant.shape
```

```python
ant.head()
```

```python
merged = df.merge(ant, left_on="name", right_on="file")
merged.shape
```

```python
merged.head()
```

```python
merged.to_csv(DATA_DIR / "ant" / "ant_metadata.csv", index=False)
```
